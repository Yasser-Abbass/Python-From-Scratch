```python
api_key = 'Please Enter you key here'
```

```python
code = """
x = 10
y = 15
print(x+y)
"""

data = {
    'api_dev_key': api_key,
    'api_option': 'paste',
    'api_paste_code': code,
    'api_paste_format': 'python'
}
response = requests.post('http://pastebin.com/api/api_post.php', data=data)
```

```python
print(response.text)
```

```python
response = requests.get('https://unsplash.com/photos/oR0uERTVyD0/download?force=true')
```

```python
print(response.url)
```

```python
fname = response.url.split('dl=')[1]
```

```python
print(fname)
```

```python
with open(fname, 'wb') as fs:
    fs.write(response.content)
```
